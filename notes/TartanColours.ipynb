```python
from PIL import ImageColor


def colour_table():
    for line in colours.split("\n"):
        yield line


ct = colour_table()

in_table = True
table = []
while in_table:
    try:
        line1 = ct.__next__()
        line2 = ct.__next__()
        line3 = ct.__next__()
        line4 = ct.__next__()
        l1 = line1.split(" ")
        ashex = l1[-2]
        r = l1[-1]
        name = " ".join(l1[:-2])
        g = line2
        b = line3
        short_code = line4
        print(f"Colour {name} {ashex} {short_code}")
        # Check that the decimal code is the same as the hex code
        rgb = ImageColor.getcolor(ashex, "RGB")
        assert int(r) == int(rgb[0]), f"r = {r} and rgb = {rgb}"
        assert int(g) == int(rgb[1]), f"g = {g} and rgb = {rgb}"
        assert int(b) == int(rgb[2]), f"b = {b} and rgb = {rgb}"
        table.append((name, ashex, short_code))
    except StopIteration:
        in_table = False
```
